```python
from collections import Counter
import numpy as np
import os
import pandas as pd
import random
from sklearn.linear_model import LogisticRegression
import torch.nn as nn
from sklearn.model_selection import train_test_split
from tqdm import tqdm_notebook
```

Basically just using this set up: https://github.com/ThilinaRajapakse/simpletransformers#setup

```python
#Data Processing

df = pd.read_csv("en_dup.csv")

df.label = pd.Categorical(df.label)
df['labels'] = df.label.cat.codes
df.head()
```

```python
train_df = df.head(1000)
train_df = pd.DataFrame({'text': train_df['content'], 'labels': train_df['labels'].astype(int)})
train_df.head()
test_df = df.tail(500)
test_df = pd.DataFrame({'text': test_df['content'], 'labels': test_df['labels'].astype(int)})
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import logging

model_args = ClassificationArgs(num_train_epochs=1,overwrite_output_dir=True)

# Create a ClassificationModel
model = ClassificationModel(
    "roberta", "roberta-base", args=model_args, use_cuda=False, num_labels=3
)

# Train the model
model.train_model(train_df)
```

```python

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(test_df, )

```
